```python
#Task 1

# Sample temperature list
temperatures = [
    8.2, 17.4, 14.1, 7.9, 18.0, 13.5, 9.0, 17.8, 13.0, 8.5,
    16.5, 12.9, 7.7, 17.2, 13.3, 8.4, 16.7, 14.0, 9.5, 18.3,
    13.4, 8.1, 17.9, 14.2, 7.6, 17.0, 12.8, 8.0, 16.8, 13.7,
    7.8, 17.5, 13.6, 8.7, 17.1, 13.8, 9.2, 18.1, 13.9, 8.3,
    16.4, 12.7, 8.9, 18.2, 13.1, 7.8, 16.6, 12.5
]

# Create empty lists for classifications
cold = []        # Temperatures below 10°C
mild = []        # Temperatures between 10°C and 15°C
comfortable = [] # Temperatures between 15°C and 20°C

# Iterate over the temperatures and classify them
for temp in temperatures:
    if temp < 10:
        cold.append(temp)
    elif 10 <= temp < 15:
        mild.append(temp)
    elif 15 <= temp <= 20:
        comfortable.append(temp)

# Print the classified lists
print("Cold temperatures:", cold)
print("Mild temperatures:", mild)
print("Comfortable temperatures:", comfortable)

#Task 2

# Using the lists classified earlier
cold_count = len(cold)           # Count items in the cold list
mild_count = len(mild)           # Count items in the mild list
comfortable_count = len(comfortable) # Count items in the comfortable list

# Printing the results
print("Number of times it was cold:", cold_count)
print("Number of times it was mild:", mild_count)
print("Number of times it was comfortable:", comfortable_count)

#Task 3

# Formula: Fahrenheit = (Celsius × 9/5) + 32

# Initialize an empty list for Fahrenheit values
temperatures_fahrenheit = []

# Iterate over the Celsius temperatures and convert them to Fahrenheit
for temp in temperatures:
    fahrenheit = (temp * 9/5) + 32
    temperatures_fahrenheit.append(fahrenheit)

# Print the converted Fahrenheit values
print("Temperatures in Fahrenheit:", temperatures_fahrenheit)

#Task 4

import matplotlib.pyplot as plt  # For optional plotting

# 1. Create empty lists for night, evening, and day temperatures
night_temps = []
evening_temps = []
day_temps = []

# 2. Assign values to time-of-day lists based on their positions
for i in range(len(temperatures)):
    if i % 3 == 0:  # Night: 00-08
        night_temps.append(temperatures[i])
    elif i % 3 == 1:  # Evening: 08-16
        evening_temps.append(temperatures[i])
    elif i % 3 == 2:  # Day: 16-24
        day_temps.append(temperatures[i])

# 3. Calculate the average day-time temperature
average_day_temp = sum(day_temps) / len(day_temps)
print("Average Day-Time Temperature:", round(average_day_temp, 2), "°C")

# 4. (Optional) Plot "day vs. temperature"
days = list(range(1, len(day_temps) + 1))  # Generate day indices
plt.plot(days, day_temps, marker='o', label='Day Temperatures', color='orange')
plt.title('Day vs. Temperature')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Task 1

def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):  # Check if the element is a list
            total += sum_nested_list(element)  # Recursively sum the nested list
        else:
            total += element  # Add the number to the total
    return total

# Test cases
nested_list_1 = [1, [2, [3, 4], 5], 6, [7, 8]]
nested_list_2 = [1, [2, 3], [4, [5]]]

print("Sum of nested_list_1:", sum_nested_list(nested_list_1))  
print("Sum of nested_list_2:", sum_nested_list(nested_list_2)) 

#Task 2

def generate_permutations(s):
    def backtrack(path, remaining, results):
        if not remaining:
            results.add("".join(path))  # Add to a set to ensure uniqueness
            return
        for i, char in enumerate(remaining):
            backtrack(path + [char], remaining[:i] + remaining[i+1:], results)
    
    results = set()
    backtrack([], list(s), results)
    return sorted(results)  # Sort the results before returning

# Test cases
print("Permutations of 'abc':", generate_permutations("abc"))  
print("Permutations of 'aab':", generate_permutations("aab"))  

#Task 3

def calculate_directory_size(directory):
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):  # If the value is a dictionary, it's a subdirectory
            total_size += calculate_directory_size(value)  # Recursively calculate its size
        else:  # If it's not a dictionary, it's a file with a size
            total_size += value
    return total_size

# Test case
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

print("Total size of directory:", calculate_directory_size(directory_structure))  
```

```python
#Task 1

def min_coins(coins, amount):
    # Initialize a DP array with a high value (infinity) for all amounts
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # Base case: 0 coins needed to make amount 0
    
    # Build the DP table
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    
    # Return the result for the target amount
    return dp[amount] if dp[amount] != float('inf') else -1

# Testing the function
coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)
print(f"Minimum coins required to make {amount} using {coins}: {result}")

#Task 2

def longest_common_subsequence(s1, s2):
    # Lengths of the two strings
    n, m = len(s1), len(s2)
    
    # Initialize a 2D DP array with dimensions (n+1) x (m+1)
    dp = [[0] * (m + 1) for _ in range(n + 1)]
    
    # Fill the DP table
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1  # Characters match
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])  # Take the best option
    
    # The value in dp[n][m] is the length of the LCS
    return dp[n][m]

# Testing the function
s1 = "abcde"
s2 = "ace"
result = longest_common_subsequence(s1, s2)
print(f"Length of LCS of '{s1}' and '{s2}': {result}")

#Task 3

def knapsack(weights, values, capacity):
    n = len(weights)
    
    # Initialize a DP table with dimensions (n+1) x (capacity+1)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    
    # Fill the DP table
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i - 1] <= w:
                # Include the item or exclude it
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                # Exclude the item
                dp[i][w] = dp[i - 1][w]
    
    # The maximum value is in the bottom-right cell of the table
    return dp[n][capacity]

# Testing the function
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
result = knapsack(weights, values, capacity)
print(f"Maximum value with weights {weights} and values {values} for capacity {capacity}: {result}")

```
